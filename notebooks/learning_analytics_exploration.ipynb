# Learning Analytics Exploration Notebook

This notebook belongs to the approved project architecture for **Automated Learning Analytics and Personalized Feedback System for Mathematics Education**. It is intended for exploratory analysis of student learning events before those insights are exposed through Flask microservices and the React 18 visualization layer.

```python
import pandas as pd

events = pd.DataFrame([
    {"student_id": "stu-001", "topic": "Algebra", "score_pct": 88, "time_spent_min": 16},
    {"student_id": "stu-001", "topic": "Calculus", "score_pct": 62, "time_spent_min": 26},
    {"student_id": "stu-002", "topic": "Trigonometry", "score_pct": 49, "time_spent_min": 21},
    {"student_id": "stu-003", "topic": "Calculus", "score_pct": 86, "time_spent_min": 20}
])

events
```

```python
topic_summary = (
    events.groupby("topic", as_index=False)
    .agg(avg_score=("score_pct", "mean"), avg_time=("time_spent_min", "mean"))
    .sort_values("avg_score")
)

topic_summary
```
